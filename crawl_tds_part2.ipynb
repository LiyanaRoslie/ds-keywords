## Crawl TDS Articles Content

### Define a few user agents inside a list 

```python
user_agent_list = [
    'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.1.1 Safari/605.1.15',
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:77.0) Gecko/20100101 Firefox/77.0',
    'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
    'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:77.0) Gecko/20100101 Firefox/77.0',
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
]
```

### Scrape articles content and store in a list

```python
# Define list to store articles content
stories_data = []

# Define total number of articles records or rows
total_article_records = len(tds_articles_df['url'])
# Define the number of article's content to scrape during each interval
interval = 100
# Define the index for the start record at 22444
start_article_record = 22444

while start_article_record < total_article_records:
    checkpoint = min(total_article_records, start_article_record + interval)
    # Select random user agent to use for each interval
    user_agent = random.choice(user_agent_list)
    headers = {'User-Agent': user_agent}
    print("Headers:", headers)
    
    for article_record_num in range(start_article_record, checkpoint):
        print("Article record number:", article_record_num)
        print("URL record:", tds_articles_df['url'][article_record_num])
        
        story_page = requests.get(tds_articles_df['url'][article_record_num], headers = headers)
        story_soup = BeautifulSoup(story_page.text, 'html.parser')
        
        sections = story_soup.find_all('section')
    
        story_paragraphs = []
    
        # parse all paragraphs in all sections
        for section in sections:
            paragraphs = section.find_all('p')
            for paragraph in paragraphs:
                story_paragraphs.append(paragraph.text)
            
        number_paragraphs = len(story_paragraphs)
        
        # concatenate items in a list to a single string
        stories_data.append(' '.join(story_paragraphs))

    start_article_record = checkpoint
    # introduce a random delay after each interval
    random_delay = random.randint(1, 10)
    print("Random Delay:", random_delay, "second(s)\n")
    time.sleep(random_delay)
```

### Convert from index 22444 list into dataframe inside column called content

```python
article_content_df = DataFrame (stories_data, columns=['content'])
article_content_df.index += 22444
article_content_df.head()
```

### Join content dataframe to the main article dataframe

```python
articles_df2 = tds_articles_df.join(article_content_df)
articles_df2.tail()
```
